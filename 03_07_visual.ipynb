### Parallel Coordinates, Numerical python, $ \ldots $.

```python
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

from pandas.plotting import parallel_coordinates

from mpl_toolkits.mplot3d import Axes3D
```

```python
np.random.seed(1234)
```

```python
from sklearn import datasets
```

```python
data = datasets.load_iris()
data_frame = pd.DataFrame(np.column_stack([data.data, data.target]), columns=data.feature_names+['Value'])
```

```python
data_frame['Name'] = data_frame['Value'].map({0.0:data.target_names[0], 1.0:data.target_names[1], 2.0:data.target_names[2]})
```

```python
data_frame.head()
```

```python
plt.figure(figsize=(20,15))
parallel_coordinates(data_frame, 'Name');
```

```python
fig = plt.figure(figsize=(10,10))
ax = Axes3D(fig)

ax.scatter(data_frame.iloc[:,0], data_frame.iloc[:,1], data_frame.iloc[:,4])
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Species');
```

```python
fig = plt.figure(figsize=(10,10))
ax = Axes3D(fig)

ax.scatter(data_frame.iloc[:,2], data_frame.iloc[:,3], data_frame.iloc[:,4])
ax.set_xlabel('Petal Length')
ax.set_ylabel('Petal Width')
ax.set_zlabel('Species');
```

```python
fig = plt.figure(figsize=(10,10))
ax = Axes3D(fig)

ax.scatter(data_frame.iloc[:,0], data_frame.iloc[:,2], data_frame.iloc[:,4])
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Petal Length')
ax.set_zlabel('Species');
```

```python
fig = plt.figure(figsize=(10,10))
ax = Axes3D(fig)

ax.scatter(data_frame.iloc[:,1], data_frame.iloc[:,3], data_frame.iloc[:,4])
ax.set_xlabel('Sepal Width')
ax.set_ylabel('Petal Width')
ax.set_zlabel('Species');
```

```python
fig = plt.figure(figsize=(10,10))
ax = Axes3D(fig)

ax.scatter(data_frame.iloc[:,0], data_frame.iloc[:,1], data_frame.iloc[:,2])
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length');
```
